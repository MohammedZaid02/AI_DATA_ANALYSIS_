```python
# Styling with Matplotlib - Line Plot
# Objective: Create a line plot with customized styles using Matplotlib.
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x + np.pi/4)

# Create the plot
plt.figure(figsize=(10, 6)) # Adjust figure size

# Plotting the first line with customizations
plt.plot(x, y1,
         color='dodgerblue',      # Line color
         linestyle='-',           # Line style (solid)
         linewidth=2,             # Line width
         marker='o',              # Marker style (circle)
         markersize=5,            # Marker size
         markerfacecolor='lightblue', # Marker fill color
         markeredgecolor='black', # Marker edge color
         label='sin(x)')          # Label for the legend

# Plotting the second line with customizations
plt.plot(x, y2,
         color='orangered',       # Line color
         linestyle='--',          # Line style (dashed)
         linewidth=2,             # Line width
         marker='^',              # Marker style (triangle up)
         markersize=6,            # Marker size
         markerfacecolor='lightsalmon',
         markeredgecolor='darkred',
         label='cos(x)')

# Plotting the third line with different customizations
plt.plot(x, y3,
         color='green',           # Line color
         linestyle=':',           # Line style (dotted)
         linewidth=2.5,           # Line width
         marker='s',              # Marker style (square)
         markersize=5,
         label='sin(x + π/4)')

# Adding a title
plt.title('Customized Line Plot Example 🎨', fontsize=16)

# Adding labels to the axes
plt.xlabel('X-axis Label (Units)', fontsize=12)
plt.ylabel('Y-axis Label (Values)', fontsize=12)

# Adding a legend
# loc='best' tries to find the best position for the legend
plt.legend(loc='lower left', fontsize=10, shadow=True, frameon=True, facecolor='lightyellow')

# Adding a grid for better readability
plt.grid(True, linestyle='--', alpha=0.7) # alpha controls transparency

# Customizing tick parameters (optional)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Adding a text annotation (optional)
plt.text(6, 0.5, 'Interesting Point!', fontsize=10, color='purple')

# Display the plot
plt.show()
```
