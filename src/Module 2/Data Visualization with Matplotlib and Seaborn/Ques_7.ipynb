```python
# Advanced Scatter Plot with Seaborn
# Description: Using Seaborn, enhance the scatter plot for height vs. weight by adding size for points based
# on data: Size [10, 15, 13, 18, 12].

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
heights = [160, 170, 165, 180, 175]        # in centimeters
weights = [55, 65, 60, 80, 76]             # in kilograms
sizes = [10, 15, 13, 18, 12]               # custom size for each point

# Create a DataFrame
data = pd.DataFrame({
    'Height': heights,
    'Weight': weights,
    'Size': sizes
})

# Create scatter plot using Seaborn with variable sizes
plt.figure(figsize=(8, 5))
sns.scatterplot(
    x='Height',
    y='Weight',
    size='Size',               # Use size for point scaling
    data=data,
    hue='Size',                # Optional: color by size
    palette='viridis',
    sizes=(50, 300),           # Scale min/max marker size
    legend=False,              # Hide legend if not needed
    edgecolor='black'
)

# Add labels and title
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Advanced Height vs Weight Scatter Plot')

plt.tight_layout()
plt.show()
```
