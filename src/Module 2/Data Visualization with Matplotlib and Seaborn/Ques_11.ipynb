```python
# Scatter Plot of Variable Relationships
# Description: Use Seaborn to create a scatter plot showing the relationship between two
# variables.

import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
tips = sns.load_dataset('tips')

# Create a scatter plot: total_bill vs tip
plt.figure(figsize=(8, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', color='blue', edgecolor='black')

# Add labels and title
plt.title('Scatter Plot: Total Bill vs Tip')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')

# Show the plot
plt.tight_layout()
plt.show()
```
