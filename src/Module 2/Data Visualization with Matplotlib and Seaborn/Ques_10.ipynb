```python
# Advanced Seaborn Plot with DataFrame
# Description: Create a Seaborn plot using a DataFrame that visualizes two features in a single plot. Use tips
# dataset available in Seaborn's sample datasets.

import seaborn as sns
import matplotlib.pyplot as plt

# Load sample 'tips' dataset
tips = sns.load_dataset('tips')

# Create scatter plot: total_bill vs tip, colored by sex
plt.figure(figsize=(8, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker', size='size', palette='Set2', sizes=(20, 200))

# Add labels and title
plt.title('Total Bill vs Tip by Gender and Smoking Status')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')

# Show plot
plt.tight_layout()
plt.show()
```
