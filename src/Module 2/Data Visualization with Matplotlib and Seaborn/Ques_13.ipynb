```python
# Customizing Plots with Seaborn - Boxplot
# Objective: Customize a boxplot with Seaborn to show the distribution of petal lengths by
# species.
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create a customized boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(
    x="species", 
    y="petal_length", 
    data=iris, 
    palette="pastel", 
    linewidth=2, 
    fliersize=5
)

# Add plot title and axis labels
plt.title("Distribution of Petal Lengths by Species", fontsize=14)
plt.xlabel("Species", fontsize=12)
plt.ylabel("Petal Length (cm)", fontsize=12)

# Display the plot
plt.tight_layout()
plt.show()
```
