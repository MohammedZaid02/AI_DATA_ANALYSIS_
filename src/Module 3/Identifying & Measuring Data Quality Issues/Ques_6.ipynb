```python
# Question: Evaluating Data Distribution
# Description: Analyze the distribution of a numeric column using histograms and descriptive statistics.

import pandas as pd
import matplotlib.pyplot as plt

def analyze_distribution(data, column_name):
    """
    Analyze the distribution of a numeric column:
    - Prints descriptive statistics (mean, median, mode, std, min, max)
    - Plots a histogram
    
    Parameters:
    - data: pandas DataFrame
    - column_name: str, name of the numeric column to analyze
    """
    if column_name not in data.columns:
        print(f"Column '{column_name}' not found in the dataset.")
        return
    
    col = data[column_name].dropna()
    
    if col.empty:
        print(f"No data available in column '{column_name}'.")
        return
    
    # Descriptive statistics
    mean = col.mean()
    median = col.median()
    mode = col.mode()
    mode_val = mode.iloc[0] if not mode.empty else None
    std = col.std()
    min_val = col.min()
    max_val = col.max()
    
    print(f"Descriptive Statistics for '{column_name}':")
    print(f"Mean: {mean}")
    print(f"Median: {median}")
    print(f"Mode: {mode_val}")
    print(f"Standard Deviation: {std}")
    print(f"Min: {min_val}")
    print(f"Max: {max_val}")
    
    # Histogram
    plt.hist(col, bins=10, edgecolor='black', alpha=0.7)
    plt.title(f"Histogram of '{column_name}'")
    plt.xlabel(column_name)
    plt.ylabel("Frequency")
    plt.grid(axis='y', alpha=0.75)
    plt.show()

# Example usage:
if __name__ == "__main__":
    sample_data = pd.DataFrame({
        'age': [23, 45, 31, 35, 40, 35, 23, 40, 45, 50, 40, 31, 31, 23, 40]
    })
    
    analyze_distribution(sample_data, 'age')
```
